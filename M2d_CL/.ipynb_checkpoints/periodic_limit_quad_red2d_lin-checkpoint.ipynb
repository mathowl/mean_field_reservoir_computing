```python
import numpy as np
import matplotlib.pyplot as plt
import random
import math
import shelve
import string
import numba as nb
import timeit
import matplotlib.cm as cm
from matplotlib.colors import Normalize
from matplotlib import colors
from scipy.stats import cauchy
from scipy.signal import savgol_filter
from tqdm import tqdm
```

```python

@nb.njit()
def flimit(rho,psi,p,w,u):
    # Delta =1
    K,F = p[0], p[1]
    
    drho = np.dtype('float64').type(-rho + (K/4)*(1-rho**2)*(rho+ rho*np.cos(2*psi)) + (F/2)*(1-rho**2)*np.cos(psi-u))
    dpsi = np.dtype('float64').type(w - (K/4)*(1+rho**2)*(np.sin(2*psi)) - (F/2)*(1/rho+rho)*np.sin(psi-u))
    
    return drho , dpsi
```

```python
@nb.njit()
def NMSE(yeval,ydata):
    
    nmse = 0
    
    for j in nb.prange(yeval.shape[0]):
        nmse += (ydata[j]-yeval[j])**2
    nmse=nmse/numba_norm(ydata[:])**2
    return nmse

@nb.njit() 
def numba_norm(a):
    n = a.shape[0]
    norm = 0
    for i in range(n):
        norm += a[i] * a[i]
    return np.sqrt(norm)
```

```python
@nb.njit()
def flimit_nou(rho,psi,p,w,Wout):
    # Delta =1
    K,F = p[0], p[1]
    
    Psi = np.ones(2,dtype='float64')
    Psi[1] = psi
    
    Upred = np.dtype('float64').type(0)

    
    for i in range(3):
        Upred += Wout[i]*Psi[i]

    
    drho = np.dtype('float64').type(-rho + (K/4)*(1-rho**2)*(rho+ rho*np.cos(2*psi)) + (F/2)*(1-rho**2)*np.cos(psi- Upred  ))
    dpsi = np.dtype('float64').type( w - (K/4)*(1+rho**2)*(rho*np.sin(2*psi)) - (F/2)*(1/rho+rho)*np.sin(psi -  Upred   ))

    return drho , dpsi

@nb.njit()
def frk4_step_nou(rho,psi,p,h,w,Wout):

    krho1, kpsi1 = flimit_nou(rho,psi,p,w,Wout)

    krho2, kpsi2 = flimit_nou(rho + 0.5*h*krho1,psi + 0.5*h*kpsi1,p,w,Wout)

    krho3, kpsi3 = flimit_nou(rho + 0.5*h*krho2 ,psi + 0.5*h*kpsi2 ,p,w,Wout)

    krho4, kpsi4 = flimit_nou(rho + h*krho3,psi + h*kpsi3 ,p,w,Wout)

    yrho = rho + h*(1.0 / 6.0)*(krho1 + 2 * krho2 + 2 * krho3 + krho4)
    ypsi = psi + h*(1.0 / 6.0)*(kpsi1 + 2 * kpsi2 + 2 * kpsi3 + kpsi4)

    return yrho, ypsi

@nb.njit()
def frk4_step(rho,psi,u0,umid,u1,p,h,w):

    krho1, kpsi1 = flimit(rho,psi,p,w,u0)

    krho2, kpsi2 = flimit(rho + 0.5*h*krho1,psi + 0.5*h*kpsi1,p,w,umid)

    krho3, kpsi3 = flimit(rho + 0.5*h*krho2 ,psi + 0.5*h*kpsi2 ,p,w,umid)

    krho4, kpsi4 = flimit(rho + h*krho3,psi + h*kpsi3 ,p,w,u1)

    yrho = rho + h*(1.0 / 6.0)*(krho1 + 2 * krho2 + 2 * krho3 + krho4)
    ypsi = psi + h*(1.0 / 6.0)*(kpsi1 + 2 * kpsi2 + 2 * kpsi3 + kpsi4)

    return yrho, ypsi

@nb.njit()
def solver(rho0,psi0,U1,U2,U3,p,h,w,steps =1000):
    Yrho = np.empty(steps, dtype = 'float64')
    Ypsi = np.empty(steps, dtype = 'float64')
    for i in nb.prange(steps):
        yrho, ypsi = frk4_step(rho0,psi0,U1[i],U2[i],U3[i],p,h,w) 
        Yrho[i] = yrho
        Ypsi[i] = ypsi
        rho0 = yrho
        psi0 = ypsi
    return Yrho, Ypsi

@nb.njit()
def solver_nou(rho0,psi0,Wout,p,h,w,steps =1000):
    Yrho = np.empty(steps, dtype = 'float64')
    Ypsi = np.empty(steps, dtype = 'float64')
    Upred = np.zeros(steps, dtype = 'float64')
    
    for i in nb.prange(steps):
        yrho, ypsi = frk4_step_nou(rho0,psi0,p,h,w,Wout) 
        Yrho[i] = yrho
        Ypsi[i] = ypsi
        rho0 = yrho
        psi0 = ypsi

        Psi = np.ones(2,dtype='float64')
        Psi[1] = ypsi
        
        for j in range(2):
            Upred[i] += Wout[j]*Psi[j]
    return Yrho, Ypsi, Upred

def udata(h,v1=0.1,M=100_000):
    Udata = v1*np.arange(0,M)*h/2
    return Udata


```

```python

h=1e-3
v1 = 2

U_input = udata(h,v1=v1,M=1_000_000)

U0 = U_input[::2]
Umid =U_input[1::2]
U1 =U_input[2::2]

initlen=25_000
trainlen=25_000
testlen = 25_000

reg=1e-5

Fh =  151
Kh =  251

F_list = [] 
K_list = []
train_list = []
test_list = []
for F in tqdm(np.linspace(0,1.85,Fh)):
    for K in np.linspace(1.75,7,Kh):
        p = np.array([K,F],dtype='float64')
        w= np.dtype('float64').type(1)
        rho0=np.dtype('float64').type(0.5)
        psi0=np.dtype('float64').type(0.5)
        
        Yrho, Ypsi = solver(rho0,psi0,U0,Umid,U1,p,h,w,steps =initlen+trainlen)
        Ypsi_sub = Ypsi[initlen:]
        X_state=np.concatenate((np.expand_dims(np.ones(len(Ypsi_sub)),axis=0),\
                                      np.expand_dims(Ypsi_sub,axis=0),\
                                    np.expand_dims(np.sin(2*Ypsi_sub),axis=0)),axis=0)
        Yt = (U0[initlen:Ypsi.shape[0]])  

        Wout = np.linalg.solve( np.matmul(X_state,X_state.T) + reg*np.eye(X_state.shape[0]) , np.matmul(X_state,Yt) ).T
        np.linalg.norm(np.abs(np.matmul(Wout,X_state)-Yt.T),2)

        Utrain = np.matmul(Wout,X_state).T
        
        Yrho_new, Ypsi_new,Upred = solver_nou(Yrho[-1],Ypsi[-1],Wout,p,h,w,steps =testlen)
        test_NMSE = NMSE(Upred, U0[initlen+trainlen:initlen+trainlen+testlen ])
        train_NMSE = NMSE(Utrain, U0[initlen:initlen+trainlen ])
        
        F_list.append(F)
        K_list.append(K)
        train_list.append(train_NMSE)
        test_list.append(test_NMSE)
        #print(f'F{F}:  | K:{K} | train_NMSE:{train_NMSE} | test_NMSE:{test_NMSE}')
        
```

```python
with shelve.open(f'bifu_limi2dred.shelve', 'c') as shelf:
    shelf['F'] = F_list
    shelf['K'] = K_list
    shelf['train'] = train_NMSE
    shelf['test'] = test_NMSE
```
