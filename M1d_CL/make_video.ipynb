```python
import numpy as np
import matplotlib.pyplot as plt
import shelve
import numba as nb
import math


from PIL import Image 
from pathlib import Path 
import os 
import imageio 
from moviepy import editor 
import moviepy as mp
```

```python
name_list =[]
for i in range(0,200):
    name_list.append(f'{1001+i}.png')
```

```python
k=1
video_path = os.path.join(os.getcwd(), "video") 
images_path = os.path.join(os.getcwd(), f"Pictures\\VIDEO{k}") 


list_of_images = []

for image_file in name_list: 
    if image_file.endswith('.png'): 
        image_path = os.path.join(images_path, image_file) 
        image = Image.open(image_path)
        list_of_images.append(image) 
  
```

```python
k=1
imageio.mimsave(f'VIDEO{k}.gif', list_of_images) 
```

```python

clip = editor.VideoFileClip(f"VIDEO{k}.gif")
clip.write_videofile(f"VIDEO{k}.mp4", codec="libx264")
```
