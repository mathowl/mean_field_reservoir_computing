```python
import numpy as np
import matplotlib.pyplot as plt
import random
import math
import shelve
import string
import numba as nb
import timeit
import matplotlib.cm as cm
from matplotlib.colors import Normalize
from matplotlib import colors
from scipy import optimize

```

```python
def train_solve(y0,nsteps,uvec,Omega,K,F,h):
    X = np.empty((nsteps,2), dtype = 'float64')
    X[0,:] = y0
    for i in range(nsteps-1):
        u0 = uvec[2*i]
        umid = uvec[2*i+1]
        u1 = uvec[2*i+2]
        X[i+1,:] = rk4_train(X[i,:],u0,umid,u1,Omega,K,F,h)
    return X
    
def test_solve(y0, nsteps,Psi0,Omega,K,F,h):
    X = np.empty((nsteps,2), dtype = 'float64')
    X[0,:] = y0
    for i in range(nsteps-1):
        X[i+1,:] = rk4_test(X[i,:],Psi0,Omega,K,F,h)
    return X


@nb.njit()
def frk4_step(y,h,K,F,Omeg):
    yout = np.empty((2),dtype='float64')
    k1 = h * f( y, K,F,Omeg)
    k2 = h * f(y + 0.5 * k1, K,F,Omeg)
    k3 = h * f( y + 0.5 * k2, K,F,Omeg)
    k4 = h * f(y + k3, K,F,Omeg)
    y = y + (1.0 / 6.0)*(k1 + 2 * k2 + 2 * k3 + k4)
    yout[0] = y[0]
    yout[1] = (y[1]  ) 
    return yout


@nb.njit()
def f(y,K,F,Omeg):
    yout = np.empty((2),dtype='float64')
    yout[0] = (-K*y[0]**3+ 2*Omeg*y[1] + F*(1-y[0]**2+y[1]**2) + y[0]*(-2 + K - K*y[1]**2))/2
    yout[1] = -1*Omeg*y[0]- y[1]*(2+2*F*y[0]+K*(-1 + y[0]**2+y[1]**2))/2
    return yout


@nb.njit()
def jacob(y,K,F,Omega):
    yout = np.empty((2,2),dtype='float64')
    
    yout[0,0] =  (-2+K-2*F*y[0]-3*K*y[0]**2-K*y[1]**2) /2
    yout[1,0] = Omega+(F-K*y[0]) * y[1]
    yout[0,1] = -Omega-(F+K *y[0])*y[1]
    yout[1,1] = (-2+K-2*F * y[0]-K * y[0]**2-3*K * y[1]**2)/2
    
    return yout

@nb.njit()
def deti(A):
    return A[0,0]*A[1,1]-A[1,0]*A[0,1]

@nb.njit()
def traci(A):
    return A[0,0]+A[1,1]



@nb.njit()
def f0(y,K,F,Omeg):
    yout = np.empty((2),dtype='float64')
    yout[0] = (-K*y[0]**3+ 2*Omeg*y[1] + F*(1-y[0]**2+y[1]**2) + y[0]*(-2 + K - K*y[1]**2))/2
    yout[1] = -1*Omeg*y[0]- y[1]*(2+2*F*y[0]+K*(-1 + y[0]**2+y[1]**2))/2
    return yout

def label_func(eigs):
    label_list = ['attracting', 'repelling', 'saddle', 'attracting - complex', 'repelling - complex']
    eig_real_sort = np.sort(eig.real)
    eig_complex = np.any(eig.imag != 0)
    if eig_real_sort[0] <0 and eig_real_sort[1] <0 and eig_complex == False:
        return 'attracting'
    elif eig_real_sort[0] <0 and eig_real_sort[1] <0 and eig_complex == True:
        return 'attracting - complex'
    elif eig_real_sort[0] >0 and eig_real_sort[1] >0 and eig_complex == False:
        return 'repelling'
    elif eig_real_sort[0] >0 and eig_real_sort[1] >0 and eig_complex == True:
        return 'repelling - complex'
    elif eig_real_sort[0] <0 and eig_real_sort[1] >0:
        return 'saddle'
    else:
        return 'undefined'
```

```python
def f_train(xvec,u,omega,K,F):
    x=xvec[0]
    y=xvec[1]
    xnew = (1/2)*(-2*x+K*x-K* x**3-2*omega*y-K* x * y**2+F*(1-x**2+y**2)*np.cos(u)-2*F*x*y*np.sin(u))
    
    ynew = (1/2)*(2*omega*x - 2*y + K*y - K * x**2 *y - K * y **3 - 2 * F * x * y * np.cos(u) + 
   F * (1 + x**2 - y**2)* np.sin(u))
    return np.array([xnew,ynew])

def f_test(xvec,Psi0,omega,K,F):
    x=xvec[0]
    y=xvec[1]
    
    xnew = (1/(2*np.sqrt( x**2 + y**2)))*(-np.sqrt(x**2 + y**2) *(K*x**3 + 2*omega*y + x*(2 + K*(-1 + y**2))) -F*x*(-1 + x**2 + y**2)*np.cos(Psi0) + F*y*(1 + x**2 + y**2)*np.sin(Psi0))
    ynew = (1/(2*np.sqrt(x**2+y**2)))*(np.sqrt(x**2+y**2)*(2*omega*x-y*(2+K*(-1+x**2+y**2)))
                            -F*y*(-1+x**2+y**2)*np.cos(Psi0)-F*x*(1+x**2+y**2)*np.sin(Psi0))
    
    return np.array([xnew, ynew])

jrho = lambda K,F,z1,z2 : -1 + (K/2)*(1 - 3*(z1**2+z2**2)) - F*np.cos(np.angle(z1+1j*z2))*np.sqrt((z1**2+z2**2))
```

```python
def rk4_train(y,u0,umid,u1,Omega,K,F,h):
    yout = np.empty((2),dtype='float64')
    k1 = h * f_train( y, u0,Omega,K,F)
    k2 = h * f_train(y + 0.5 * k1, umid,Omega,K,F)
    k3 = h * f_train( y + 0.5 * k2, umid,Omega,K,F)
    k4 = h * f_train(y + k3, u1,Omega,K,F)
    y = y + (1.0 / 6.0)*(k1 + 2 * k2 + 2 * k3 + k4)
    yout[0] = y[0]
    yout[1] = y[1]   
    return yout    

def rk4_test(y,Psi0,Omega,K,F,h):
    yout = np.empty((2),dtype='float64')
    k1 = h * f_test( y, Psi0,Omega,K,F)
    k2 = h * f_test(y + 0.5 * k1, Psi0,Omega,K,F)
    k3 = h * f_test( y + 0.5 * k2, Psi0,Omega,K,F)
    k4 = h * f_test(y + k3, Psi0,Omega,K,F)
    y = y + (1.0 / 6.0)*(k1 + 2 * k2 + 2 * k3 + k4)
    yout[0] = y[0]
    yout[1] = y[1]      
    return yout
    
```

```python
f1 = lambda K,scale,vu : scale**2*np.sqrt(((K-2*scale)*(K**4-4*K**3*scale+16*K*scale*vu**2+16*scale**2*vu**2+4*K**2*(scale**2+vu**2)))/((K+2*scale)*scale**4))/(2*K)

f2 = lambda rho, scale, vu : (2*scale*(1 + rho**6 - np.sqrt(((1 + rho**2)**3*(rho**4*(1 + rho**2)**3 + ((-1 + rho**2)**5*vu**2)/scale**2))/(-1 + rho**2)**4) - 
    rho**4*(-3 + np.sqrt(((1 + rho**2)**3*(rho**4*(1 + rho**2)**3 + ((-1 + rho**2)**5*vu**2)/scale**2))/(-1 + rho**2)**4)) + 
    rho**2*(3 + 2*np.sqrt(((1 + rho**2)**3*(rho**4*(1 + rho**2)**3 + ((-1 + rho**2)**5*vu**2)/scale**2))/(-1 + rho**2)**4))))/((-1 + rho**2)**2*(1 + rho**2)**3)


f3 = lambda rho, scale, vu : (2*np.sqrt(2)*rho**2*scale*np.sqrt((rho**2 + 3*rho**4 + 3*rho**6 + rho**8 + ((-1 + rho**2)**4*vu**2)/scale**2 - 
      np.sqrt(((1 + rho**2)**3*(rho**4*(1 + rho**2)**3 + ((-1 + rho**2)**5*vu**2)/scale**2))/(-1 + rho**2)**4) + 
      2*rho**2*np.sqrt(((1 + rho**2)**3*(rho**4*(1 + rho**2)**3 + ((-1 + rho**2)**5*vu**2)/scale**2))/(-1 + rho**2)**4) - 
      rho**4*np.sqrt(((1 + rho**2)**3*(rho**4*(1 + rho**2)**3 + ((-1 + rho**2)**5*vu**2)/scale**2))/(-1 + rho**2)**4))/(1 + rho**2)**3))/(-1 + rho**2)**2

f4 = lambda rho, scale, vu : (2*scale*(1 + rho**6 + np.sqrt(((1 + rho**2)**3*(rho**4*(1 + rho**2)**3 + ((-1 + rho**2)**5*vu**2)/scale**2))/(-1 + rho**2)**4) + 
    rho**4*(3 + np.sqrt(((1 + rho**2)**3*(rho**4*(1 + rho**2)**3 + ((-1 + rho**2)**5*vu**2)/scale**2))/(-1 + rho**2)**4)) - 
    rho**2*(-3 + 2*np.sqrt(((1 + rho**2)**3*(rho**4*(1 + rho**2)**3 + ((-1 + rho**2)**5*vu**2)/scale**2))/(-1 + rho**2)**4))))/((-1 + rho**2)**2*(1 + rho**2)**3)

f5 = lambda rho, scale, vu : (2*np.sqrt(2)*rho**2*scale*np.sqrt((rho**2 + 3*rho**4 + 3*rho**6 + rho**8 + ((-1 + rho**2)**4*vu**2)/scale**2 + 
      np.sqrt(((1 + rho**2)**3*(rho**4*(1 + rho**2)**3 + ((-1 + rho**2)**5*vu**2)/scale**2))/(-1 + rho**2)**4) - 
      2*rho**2*np.sqrt(((1 + rho**2)**3*(rho**4*(1 + rho**2)**3 + ((-1 + rho**2)**5*vu**2)/scale**2))/(-1 + rho**2)**4) + 
      rho**4*np.sqrt(((1 + rho**2)**3*(rho**4*(1 + rho**2)**3 + ((-1 + rho**2)**5*vu**2)/scale**2))/(-1 + rho**2)**4))/(1 + rho**2)**3))/(-1 + rho**2)**2
ftest = lambda K1,sigma : np.sqrt(8*K1 + 52*K1**2 + 94*K1**3 + 43*K1**4 + 2*K1**5 - 8*K1*np.sqrt(1 + 4*K1) - 
        36*K1**2*np.sqrt(1 + 4*K1) - 
        38*K1**3*np.sqrt(1 + 4*K1) - 7*K1**4 *np.sqrt(1 + 4*K1) + 2*K1*sigma**2 + 
        7*K1**2* sigma**2 + 2*K1**3 *sigma**2 - 2*K1*np.sqrt(1 + 4*K1) *sigma**2 -  
        3*K1**2*np.sqrt(1 + 4*K1)* sigma**2)/np.sqrt(1 + 6 *K1 + 9*K1**2 + 
        2*K1**3 - np.sqrt(1 + 4*K1) - 4 * K1 *np.sqrt(1 + 4*K1) - 
        3*K1**2*np.sqrt(1 + 4*K1))
ftest2 = lambda K1,sigma : np.sqrt(-2+K1)*np.sqrt(K1)*sigma/(-1+K1)
```

```python

def frk4test_step(y,h,Psi0,omega,K,F):
    yout = np.empty((2),dtype='float64')
    k1 = h * f_test( y, Psi0,omega,K,F)
    k2 = h * f_test(y + 0.5 * k1, Psi0,omega,K,F)
    k3 = h * f_test( y + 0.5 * k2, Psi0,omega,K,F)
    k4 = h * f_test(y + k3, Psi0,omega,K,F)
    y = y + (1.0 / 6.0)*(k1 + 2 * k2 + 2 * k3 + k4)
    yout[0] = y[0]
    yout[1] = (y[1]  ) 
    return yout
```

```python
h=0.01
K=4
N=10
Omega=1


@nb.njit()
def f0test(y,Psi0,Omeg,K,F):
    yout = np.empty((1),dtype='float64')
    yout = -y + (K/2)*y*(1-y**2) + (F/2)*(1-y**2)*np.cos(Psi0)
    return yout

def jacobtest(y,Psi0,Omeg,K,F):
    yout = np.empty((1),dtype='float64')
    yout = -1 + (K/2)*(1-3*y**2) - (F)*(y)*np.cos(Psi0)
    return yout
```

```python
h=0.01
c =2
uvec = np.arange(0,30_000,h/2)*c
K=4
N=10
Omega=1
indo0 = 1000
for F in np.linspace(0.5,1.2,200):
    indo0 +=1
    sol_set = set()

    xv, yv = np.meshgrid(np.linspace(-1,1,10), np.linspace(-1,1,10), indexing='ij')
    for x0,y0 in zip(xv.flatten(),yv.flatten()):
        if x0**2 + y0**2 < 1.001:
            fun = lambda y : f0(y,K,F,Omega)
            jfun = lambda y : jacob(y,K,F,Omega)
            sol = optimize.root(fun, [x0, y0], jac=jfun, method='hybr',tol = 10 ** (-8))
            if sol.success == True:

                sol_set.add( (np.round(sol.x,6)[0],(np.round(sol.x,6)[1] )))
    sol_list = list(sol_set)
    eig_list = list()
    for s in sol_list:
        eigenval, _  = np.linalg.eig(jfun(np.array(s)))
        eig_list.append(eigenval)
        
    fig = plt.figure(figsize=(17,18)) 
    fig.suptitle(f'Bifurcation parameters K:{np.round(K,4)}, F:{np.round(F,4)} (black dot 1st fig)', fontsize=20)
    plt.subplots_adjust(top=0.92)
    ax1 = plt.subplot(4,3,(1,4))
    ax1.set_title('Bifurcation testing and training', fontsize=16)
    ax1.set_xlim([1.8,6])
    ax1.set_ylim([0,1.8])
    ax1.scatter(10,10,color='grey',label='success')
    ax1.scatter(10,10,color='white',label='failure')
    ax1.scatter(K,F,color='black',alpha=1)
    ax1.legend(loc='upper right')
    ax1.set_xlabel('K')
    ax1.set_ylabel('F')
    
    ax1.fill_between(full_domK , full_domF , 1.8 , color='grey')#, alpha=0.5 )
    ax1.grid()
    ax1.scatter(K,F,color='black',alpha=1)
    ax2 =  plt.subplot(4,3,2)
    xv, yv = np.meshgrid(np.linspace(-1,1,26), np.linspace(-1,1,26), indexing='ij')
    for x0,y0 in zip(xv.flatten(),yv.flatten()):
        if x0**2 + y0**2 < 1.1:
            ysol = np.zeros( (2,N))
            ysol[:,0] = [x0,y0]
            for i in range(0,N-1):
                ysol[:,i+1]=frk4_step(ysol[:,i],h,K,F,Omega)

            ax2.plot(ysol[0,:],ysol[1,:],color='black')
            ax2.plot(ysol[0,-1],ysol[1,-1],'.',color='black')
    color_dict = {'attracting' : 'blue', 'attracting - complex': 'purple',  'repelling': 'red',  'repelling - complex' : 'pink', 'saddle': 'green'} 
    maker_dict = {'attracting' : '^',  'attracting - complex': '^', 'repelling': 's', 'repelling - complex' : 's', 'saddle': 'x'}    
    

    
    for l in color_dict.keys():
        ax2.scatter(10 ,10,marker= maker_dict[l], label=f'{l}',color=color_dict[l])
    for s,eig  in zip(sol_list,eig_list ):
        label= label_func(eig)
        drho = jrho(K,F,s[0],s[1])
        detjac = deti(jacob(s,K,F,Omega))
        dettrac = traci(jacob(s,K,F,Omega))
        ax2.scatter(s[0] ,s[1],marker= maker_dict[label],color=color_dict[label],s=60)

        #plt.scatter(s[0] ,s[1],marker= 's', label=f'{label}:drho{np.round(drho,1)}:det{np.round(detjac,1)}:tr{np.round(dettrac,2)}',color=color_dict[label])
    ax2.legend(loc='upper right',framealpha=1)
    ax2.set_title('z vector field training', fontsize=16)
    ax2.set_xlim([-1,1])
    ax2.set_ylim([-1,1])
    ax2.set_xlabel('Re z')
    ax2.set_ylabel('Im z')
    ax2.set_aspect('equal')
    
    ax3 = plt.subplot(4,3,(5,6))
    n_steps=2000
    Omeg=1
    X=train_solve(np.array([0.2,0.2]),2000,uvec,Omeg,K,F,h)
    utrain = np.array([   np.angle(complex(x,y)) for x,y in zip(X[:,0],X[:,1])]) 
    w = ((uvec[::2][:2_000] % (2*np.pi))[-1] - ( utrain % (2*np.pi))[-1]) % (2*np.pi)

    Xtest = test_solve(X[-1,:], 9*n_steps,-w,Omeg,K,F,h)
    utest = np.array([   np.angle(complex(x,y)) for x,y in zip(Xtest[:,0],Xtest[:,1])]) 
    ax3.set_title('Training and testing for u', fontsize=16)
    ax3.plot(np.arange(0,20_000,1)*h, uvec[::2][:20_000] % (2*np.pi), '.', color='green', alpha=0.5,label='u(t)=ct')
    ax3.plot(np.arange(0,2_000,1)*h,(utrain+w) % (2*np.pi), '.',color='cyan',label='training $\hat{u}=Wh(\psi)$')
    ax3.plot(np.arange(2_000,20_000,1)*h, (utest + w) % (2*np.pi), '.',c='C0',label = 'testing $\hat{u}=Wh(\psi)$')
    ax3.legend(loc='upper right',framealpha=1,fontsize=12)

    ax3.set_xlim([10,30])
    ax3.set_ylim([-0.1,6.4])
    ax3.set_xlabel('t')
    Psi0= -w
    
    ax4 = plt.subplot(4,3,3)
    sol_set = set()
    xv0 = np.linspace(0,1.1,10)
    for x0 in xv0.flatten():
        fun = lambda y : f0test(y,Psi0,Omeg,K,F)
        jfun = lambda y : jacobtest(y,Psi0,Omeg,K,F)
        sol = optimize.root(fun, x0, jac=jfun, method='hybr',tol = 10 ** (-8))
        if sol.success == True and sol.x>=0 and sol.x<=1:
            sol_set.add( (np.round(sol.x,6)[0]))
    sol_list = list(sol_set)
    eig_list = list()
    for s in sol_list:
        eig_list.append(jacobtest(s,Psi0,Omeg,K,F))
    
    xv, yv = np.meshgrid(np.linspace(-1,1,26), np.linspace(-1,1,26), indexing='ij')
    for x0,y0 in zip(xv.flatten(),yv.flatten()):
        if x0**2 + y0**2 < 1.1:
            ysol = np.zeros( (2,N))
            ysol[:,0] = [x0,y0]
            for i in range(0,N-1):
                ysol[:,i+1]=frk4test_step(ysol[:,i],h,Psi0,-1*Omega,K,F)

            ax4.plot(ysol[0,:],ysol[1,:],color='black')
            ax4.plot(ysol[0,-1],ysol[1,-1],'.',color='black')
    colortest_dict = {'attracting' : 'blue',  'repelling': 'red'}


    for l in colortest_dict.keys():
        ax4.scatter(10 ,10, label=f'{l}',color=colortest_dict[l])
    for s,eig  in zip(sol_list,eig_list ):
        if eig>0:
            cir=plt.Circle((0, 0), s, color=colortest_dict['repelling'] ,fill=False)
            ax4.add_patch(cir)
        elif eig<0:
            cir=plt.Circle((0, 0), s, color=colortest_dict['attracting'] ,fill=False)
            ax4.add_patch(cir)
    ax4.legend(loc='upper right',framealpha=1)
    ax4.set_title('z vector field testing', fontsize=16)
    ax4.set_xlim([-1,1])
    ax4.set_ylim([-1,1])
    
    ax4.set_xlabel('Re z')
    ax4.set_ylabel('Im z')
    ax4.set_aspect('equal')
    fig.savefig(f'Pictures/VIDEO2/{indo0}.png',dpi=100,bbox_inches = 'tight')

    plt.close()
```
